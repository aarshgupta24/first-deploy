```python
x = db.drop('FWI',axis=1)
y= db['FWI']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
plt.figure(figsize=(12,7))
corr = x_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def coorelation(db,thres):
    col_corr=set()
    col_m = db.corr()
    for i in range(len(col_m.columns)):
        for j in range(i):
            if abs(col_m.iloc[i,j])>thres:
                col = col_m.columns[i]
                col_corr.add(col)
    return col_corr
```

```python
fea = coorelation(x_train,0.85)
```

```python
x_train.drop(fea,axis=1,inplace=True)
x_test.drop(fea,axis=1,inplace=True)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error
reg = LinearRegression()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(mae,r2)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score,mean_absolute_error
lass = Lasso(alpha=0.05725)
lass.fit(x_train,y_train)
y_pred = lass.predict(x_test)
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(mae,r2)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import LassoCV
lsscv=LassoCV(cv=5)
lsscv.fit(x_train,y_train)
```
